WHILE LOOP

```python
## While Lopp
count =0

while count<5:
    print(count)
    count = count+1
```

```python
count=0

while count%2==0:
    print(count)
```

```python
## LOOP CONTROL STATEMENTS

## break
## Exit the loop prematurely

for i in range(10):
    if i==5:
        break
    print(i)
```

```python
## continue
## continue statment and skips the current iteration
for i in range(10):
    if i==5:
        continue
    print(i)
```

```python
for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
## Pass
## The pass statement is a null operation; it does nothing.

for i in range(5):
    if i==2:
        print("The number is ", i)
        pass
    print(i)
```

```python
## Nested Loops

for i in range(3):
    for j in range(2):
        print(f"i:{i} and j;{j}")
```

```python
## Examples - calculate the sum of first N natural number using a while and for loop

num = int(input("Enter a number"))
sum=0
for i in range(num+1):
    sum = sum + i
print(sum)    
```

```python
num = int(input("Enter a number"))
sum=0

while(num):
    sum = sum +num
    num = num -1
print(sum)    
```
